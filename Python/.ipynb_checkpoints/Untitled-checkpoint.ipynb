```python
from bs4 import BeautifulSoup
import requests
for i in range (3,10):
    res = requests.get('https://hotel.travel.rakuten.co.jp/hotelinfo/plan/129475?f_teikei=&f_hizuke=&f_hak=&f_dai=japan&f_chu=tokyo&f_shou=tokyo&f_sai=E&f_tel=&f_target_flg=&f_tscm_flg=&f_p_no=&f_custom_code=&f_search_type=&f_camp_id=&f_static=1&f_rm_equip=&f_hi1={checkin}&f_tuki1=2&f_nen1=2018&f_hi2={checkout}&f_tuki2=2&f_nen2=2018&f_heya_su=1&f_otona_su=1&f_kin2=0&f_kin=&f_s1=0&f_s2=0&f_y1=0&f_y2=0&f_y3=0&f_y4=0'.format(checkin=i,checkout=i+1))
    soup = BeautifulSoup(res.content, "lxml")
    res.raise_for_status()
    price = soup.find_all(class_ ='vPrice' )
    for item in price:
        print (item.text)
    
```
